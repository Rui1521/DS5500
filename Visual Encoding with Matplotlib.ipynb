```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("Pokemon.csv")
```

## Visualizations

```python
attack_avg = data[['Type 1', 'Attack']].groupby('Type 1', as_index=False).mean()
plt.figure(figsize=(12,5))
plt.bar(attack_avg['Type 1'], attack_avg['Attack'], align='center', alpha=0.5)
y_pos = np.arange(len(attack_avg['Type 1']))
plt.xticks(y_pos, attack_avg['Type 1'])
plt.ylabel('Average Attack Strength',size = 12)
plt.xlabel('Type of Pokemon',size = 12)
plt.title('Average Attack Strength of Each Type of Pokemon',size = 15)
plt.ylim(0,120)
plt.show()
```

a. There are two attributes, one quantitive attribute and one categorical attribute. Line plots are not suitable for this because the categorical attribute, types of pokemon, has no order. The line plot would convey unnecessary implication of trend relationships, which is wrong(VAD P156). And even though pie charts and polar area charts are also used to encode one quantitative attribute and one categorical attribute, there is no part-whole relationship between two attributes. Thus, pie charts and polar area charts are not suitable for this problem neither (VAD P169). 

b. 
Advantages: The bar chart uses a line mark and encode a quantitative value attribute with one spatial position channel, which is very noticable and effective. It expresses all of, and only, the information in these two attributes(Lecture: Marks & Channel).

Disadvantages: People can easily make comparisons of two adjacent bars, but are not able to compare two bars that are not neighbors at once.

```python
plt.figure(figsize=(12,5))
attack_type = data[['Type 1', 'Attack']]
types = list(set(attack_type['Type 1']))
lst = [list(attack_type[attack_type['Type 1'] == types[i]]['Attack']) for i in range(len(types))]
plt.boxplot(lst)
plt.xticks(list(range(1,len(types)+1)), types)
plt.xlabel('Type of Pokemon',size = 12)
plt.ylabel("Attack Strength",size = 12)
plt.title("Distributions of Attack Strengths of Each Type of Pokemon",size = 15)
plt.show()
```
